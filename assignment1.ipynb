## 2. Tranformation

```python
# check unique values in Mileage to identify units and non-numeric values
data_cleaned['Mileage'].unique()
```

```python
# check in new data which car has Mileage in km/kg
data_cleaned.query('Mileage.str.contains("km/kg", na=False)', engine='python')
```

```python
# transform object to float and convert km/kg to kmpl
def convert_mileage(mileage):
    if isinstance(mileage, str):
        if 'km/kg' in mileage:
            value = float(mileage.split()[0])
            return value * 1.4  # Convert km/kg to kmpl
        elif 'kmpl' in mileage or 'km/l' in mileage:
            return float(mileage.split()[0])  # Extract value
    return None

# Apply the conversion to 'Mileage'
data_cleaned['Mileage'] = data_cleaned['Mileage'].apply(convert_mileage)
```

```python
# transform object to float 
def convert_engine(engine):
    if isinstance(engine, str):
        return float(engine.split()[0])
    return None

# Apply the conversion to 'Engine'
data_cleaned['Engine'] = data_cleaned['Engine'].apply(convert_engine)
```

```python
# function to detect outliers using IQR
def detect_outliers_iqr(data):
    q1 = data.quantile(0.25)  # First quartile (25th percentile)
    q3 = data.quantile(0.75)  # Third quartile (75th percentile)
    iqr = q3 - q1  # Interquartile range
    lower_bound = q1 - 1.5 * iqr  # Lower bound
    upper_bound = q3 + 1.5 * iqr  # Upper bound
    outliers = data[(data < lower_bound) | (data > upper_bound)]  # Identify outliers
    return outliers
```

## 4. Pearson correclation coefficient and scatter plot

```python
# this a function to remove outliers
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# remove outliers from all columns
columns_to_check = ['Car_Age', 'Kilometers_Driven', 'Mileage', 'Engine', 'Power', 'Seats', 'Price']
for col in columns_to_check:
    data_cleaned = remove_outliers(data_cleaned, col)
```

```python
# this is a function Pearson correlation coefficient calculation
# corr(x,y)= standard(x*y)/stand(x)*stand(y)
def pearson_correlation(x, y):
    mean_x, mean_y = np.mean(x), np.mean(y)
    numerator = np.sum((x - mean_x) * (y - mean_y))
    denominator = np.sqrt(np.sum((x - mean_x) ** 2) * np.sum((y - mean_y) ** 2))

    return np.nan if denominator == 0 else numerator / denominator
```

```python
# calculate Pearson correlation coefficients
correlation_results = {}
for i, col1 in enumerate(columns_to_check):
    for col2 in columns_to_check[i + 1:]:
        corr = pearson_correlation(data_cleaned[col1], data_cleaned[col2])
        correlation_results[(col1, col2)] = corr

print("Pearson Correlation Coefficients:")
for pair, corr in correlation_results.items():
    print(f"{pair}: {corr:.2f}")
```

```python
# scatter plots with correlation coefficients
def draw_scatter_plot(x, y, corr, xlabel, ylabel):
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, alpha=0.7, edgecolors='w', s=80)
    plt.title(f'Scatter Plot\nCorrelation: {corr:.2f}')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()

# draw scatter plots for each pair of columns
for (col1, col2), corr in correlation_results.items():
    draw_scatter_plot(data_cleaned[col1], data_cleaned[col2], corr, col1, col2)
```

## 5. Handling missing value

```python
data_impute = dataset.copy()
```

```python
data_impute.info()
```

```python
data_impute[['Mileage', 'Engine', 'Power', 'Seats', 'Price']].isnull().sum()
```

```python
# function to detect outliers using IQR
def detect_outliers_iqr(data):
    q1 = data.quantile(0.25)  # First quartile (25th percentile)
    q3 = data.quantile(0.75)  # Third quartile (75th percentile)
    iqr = q3 - q1  # Interquartile range
    lower_bound = q1 - 1.5 * iqr  # Lower bound
    upper_bound = q3 + 1.5 * iqr  # Upper bound
    outliers = (data < lower_bound) | (data > upper_bound)  # Identify outliers
    return outliers

# combine missing values with outliers
col_list = ["Mileage", "Engine", "Power", "Seats", "Price"]

# iterate over each column, detect outliers, and mark them as missing
for col in col_list:
    # detect outliers
    outliers = detect_outliers_iqr(data_impute[col])
    
    # replace outliers with NaN (treating them as missing values)
    data_impute.loc[outliers, col] = np.nan

# check the updated missing values (combined with treated outliers)
combined_missing_outliers = pd.DataFrame({
    "Missing Value": data_impute[col_list].isnull().sum() 
})

# updated missing values (which now includes treated outliers)
combined_missing_outliers
```

```python
data_impute[['Engine', 'Power', 'Seats']].isnull().sum()
```

```python
# Assuming 'data_impute' is your DataFrame and has already been defined
columns_to_impute = ['Engine', 'Power', 'Seats']
subclass_columns = ['Fuel_Type', 'Transmission']  # Define subclass columns

# Create a new column to track imputation status
for col in columns_to_impute:
    # Create a new column to indicate whether the value was imputed
    data_impute[f'{col}_Imputed'] = False
    
    # Impute 'Engine' and 'Seats' using mode based on subclass
    if col in ["Engine", "Seats"]:
        mode_value = data_impute.groupby(subclass_columns)[col].transform(lambda x: x.mode()[0] if not x.mode().empty else x.median())
        missing_before = data_impute[col].isnull()
        data_impute[col] = data_impute[col].fillna(mode_value)
        # Set indicator to True for imputed values
        data_impute[f'{col}_Imputed'] = missing_before != data_impute[col].isnull()

    # Impute 'Power' using mean based on subclass
    if col == "Power":
        mean_value = data_impute.groupby(subclass_columns)[col].transform(lambda x: x.mean())
        missing_before = data_impute[col].isnull()
        data_impute[col] = data_impute[col].fillna(mean_value)
        # Set indicator to True for imputed values
        data_impute[f'{col}_Imputed'] = missing_before != data_impute[col].isnull()
```

```python
# select only numeric columns for correlation
numeric_columns = data_impute.select_dtypes(include=['float64', 'int64'])

# calculate correlation for numeric columns
correlations = numeric_columns.corr()

# find columns with strong correlation to 'Mileage' and 'Price'
print("Correlation with Mileage:")
print(correlations['Mileage'].sort_values(ascending=False))

print("Correlation with Price:")
print(correlations['Price'].sort_values(ascending=False))
```

```python
from sklearn.linear_model import LinearRegression

# function to impute missing values using linear regression
def impute_with_regression(target_col, feature_cols):
    model = LinearRegression()
    non_missing_mask = data_impute[target_col].notna()
    
    # train the model on non-missing data
    model.fit(data_impute.loc[non_missing_mask, feature_cols], data_impute.loc[non_missing_mask, target_col])
    
    # predict and impute missing values
    missing_mask = data_impute[target_col].isna()
    if missing_mask.any():
        data_impute.loc[missing_mask, target_col] = model.predict(data_impute.loc[missing_mask, feature_cols])
        # add more colunms Mileage_Imputed and Price_Imputed
        data_impute[f'{target_col}_Imputed'] = missing_mask

# get correlated columns for Mileage and Price
mileage_corr_cols = correlations['Mileage'].drop('Mileage').abs().nlargest(3).index.tolist()
price_corr_cols = correlations['Price'].drop('Price').abs().nlargest(4).index.tolist()

# impute missing 'Mileage' and 'Price'
impute_with_regression('Mileage', mileage_corr_cols)
impute_with_regression('Price', price_corr_cols)
```
